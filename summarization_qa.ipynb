```python
from transformers import pipeline
with open("article.txt", "r", encoding="utf-8") as file:
    article = file.read()
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
summary = summarizer(article, max_length=200, min_length=80, do_sample=False)[0]['summary_text']
question = "What is the main humanitarian issue in Gaza?"
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
answer = qa_pipeline(question=question, context=article)['answer']

with open("output.txt", "w", encoding="utf-8") as out_file:
    out_file.write("=== Original Article ===\n")
    out_file.write(article + "\n\n")
    out_file.write("=== Summary ===\n")
    out_file.write(summary + "\n\n")
    out_file.write("=== Answer ===\n")
    out_file.write(answer)

print("Done! Check output.txt")
```
